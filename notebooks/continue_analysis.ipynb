```python
import pandas as pd

df = pd.read_json('../clean_data/data.json')

display(df.head(10))
```

```python
columns = [
    "user_agent",
    "js_fingerprint_profile_vendor",
    "tls_tls_version_record",
    "tls_tls_version_negotiated",
    "tls_ja3",
    "tls_ja3_hash",
    "tls_peetprint_hash",
    "http2_akamai_fingerprint",
    "http2_sent_frames",
    "tcpip_ip_ttl",
    "tcpip_tcp_mss",
    "tcpip_tcp_window",
    "os_prediction_highest",
    "os_prediction_platform_mismatch",
    "tls_supported_versions"
]

for col in columns:
    print(df[col].value_counts())
    print('__________________________________________________')
```

```python
one_hot = df["tls_supported_versions"].explode().reset_index(drop=True).str.get_dummies().astype(bool)
df = df.drop(columns=["tls_supported_versions"]).join(one_hot)
```

```python
columns_to_drop = [
    "tls_tls_version_record",
    "tls_tls_version_negotiated",
    "http2_sent_frames",
    "os_prediction_platform_mismatch",
]

df = df.drop(columns=columns_to_drop)

display(df.head(5))
```
